````python
# Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each method.
````

````python
import numpy as np

# Method 1: Using np.array()
array1 = np.array([[1, 2, 3], [4, 5, 6]])

print("Method 1 Output:")
print(array1)
````

````python
import numpy as np

# Method 2: Using np.zeros() and filling with values
array2 = np.zeros((2, 3), dtype=int)
array2[0, :] = [1, 2, 3]
array2[1, :] = [4, 5, 6]

print("Method 2 Output:")
print(array2)
````

````python
import numpy as np

# Method 3: Using np.full() and filling with values
array3 = np.full((2, 3), 0)
array3[0, :] = [1, 2, 3]
array3[1, :] = [4, 5, 6]

print("Method 3 Output:")
print(array3)
````

````python
Explain the following terms
- The difference in np.array, np.asarray and np.asanyarray
- The difference between Deep copy and shallow copy

Sure! Let’s break down the differences between `np.array`, `np.asarray`, and `np.asanyarray`, and then discuss the concepts of deep copy and shallow copy.

### Differences Between `np.array`, `np.asarray`, and `np.asanyarray`

1. **`np.array`**:
   - **Purpose**: Creates a new array.
   - **Behavior**: Always creates a new array object, even if the input is already an array. This means it copies the data by default, unless you explicitly pass `copy=False`.
   - **Use Case**: When you need a new array and potentially want to copy the input data.

   ```python
   import numpy as np

   arr = np.array([1, 2, 3])          # Creates a new array
   arr_from_list = np.array([1, 2, 3], dtype=np.float64)  # Creates a new array with specified dtype
   ```

2. **`np.asarray`**:
   - **Purpose**: Converts an input to an array.
   - **Behavior**: Converts the input to a NumPy array if it is not already one. If the input is already an array, it returns it as is, without copying the data unless `dtype` or `order` is specified.
   - **Use Case**: When you want to ensure the input is a NumPy array but avoid unnecessary copying.

   ```python
   import numpy as np

   arr = np.asarray([1, 2, 3])        # Converts the list to an array (no copy if input is already an array)
   arr_existing = np.asarray(np.array([1, 2, 3]))  # No copy, returns the existing array
   ```

3. **`np.asanyarray`**:
   - **Purpose**: Converts the input to an array, similar to `np.asarray`.
   - **Behavior**: Converts the input to a NumPy array but allows for certain subclasses of `np.ndarray` (e.g., masked arrays) to be preserved. It does not convert these subclasses into regular `np.ndarray`.
   - **Use Case**: When you want to preserve the input’s array subclass if applicable.

   ```python
   import numpy as np

   arr = np.asanyarray([1, 2, 3])    # Converts list to array
   arr_masked = np.ma.masked_array([1, 2, 3], mask=[0, 1, 0])
   arr_result = np.asanyarray(arr_masked)  # Preserves the masked array subclass
   ```

### Differences Between Deep Copy and Shallow Copy

1. **Shallow Copy**:
   - **Definition**: Creates a new object but does not create copies of nested objects. Instead, it inserts references to the original objects.
   - **Behavior**: Changes to the nested objects in the copied object will affect the original object.
   - **Method**: In Python, a shallow copy can be created using the `copy` module’s `copy()` function or methods like slicing (e.g., `a[:]`).

   ```python
   import copy

   original = [[1, 2, 3], [4, 5, 6]]
   shallow_copy = copy.copy(original)
   shallow_copy[0][0] = 100

   print(original)         # Output: [[100, 2, 3], [4, 5, 6]]
   print(shallow_copy)    # Output: [[100, 2, 3], [4, 5, 6]]
   ```

2. **Deep Copy**:
   - **Definition**: Creates a new object and recursively copies all objects found in the original object. This means the new object is entirely independent of the original.
   - **Behavior**: Changes to the copied object or its nested objects do not affect the original object.
   - **Method**: In Python, a deep copy can be created using the `copy` module’s `deepcopy()` function.

   ```python
   import copy

   original = [[1, 2, 3], [4, 5, 6]]
   deep_copy = copy.deepcopy(original)
   deep_copy[0][0] = 100

   print(original)         # Output: [[1, 2, 3], [4, 5, 6]]
   print(deep_copy)       # Output: [[100, 2, 3], [4, 5, 6]]
   ```
````

````python
Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

a)Extract all even integers from array.

b)Extract all odd integers from array.
````

````python
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
array = np.random.randint(1, 11, (5, 6))

print("Original Array:")
print(array)

# a) Extract all even integers from the array
even_mask = (array % 2 == 0)
even_integers = array[even_mask]

print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers from the array
odd_mask = (array % 2 != 0)
odd_integers = array[odd_mask]

print("\nOdd Integers:")
print(odd_integers)
````

````python
# Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=20, high=35, size=len(dates))
humidity = np.random.uniform(low=30, high=80, size=len(dates))

# Create the DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plotting
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis to plot Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
````

````python
# Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')

# Overlay the PDF
xmin, xmax = plt.xlim()  # Get the current x-axis limits
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
````

````python
# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)

# Determine the quadrant for each point
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant 1'
    elif x < 0 and y >= 0:
        return 'Quadrant 2'
    elif x < 0 and y < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

quadrants = np.array([get_quadrant(xi, yi) for xi, yi in zip(x, y)])

# Create a DataFrame
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': quadrants
})

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
````

````python
# With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
````

````python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display Bokeh plots inline in Jupyter notebooks
output_notebook()

# Generate data for the sine wave function
x = np.linspace(0, 4 * np.pi, 100)  # x values from 0 to 4*pi
y = np.sin(x)  # y values as sine of x

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='Sine Wave', line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3  # Make grid lines slightly transparent

# Show the plot
show(p)
````

````python
# Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.
````

````python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Display Bokeh plots inline in Jupyter notebooks
output_notebook()

# Generate random categorical data
np.random.seed(0)  # For reproducibility
categories = [f'Category {i+1}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(x_range=df['Category'], title='Random Categorical Bar Chart',
           x_axis_label='Category', y_axis_label='Value',
           width=800, height=400, toolbar_location=None, tools="")

# Add bars with color mapping
p.vbar(x='Category', top='Value', width=0.9, source=source,
       color='blue', legend_field='Value', line_color='white')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Customize grid and legend
p.grid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
````

````python
# Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
````

````python
import numpy as np
import plotly.graph_objects as go

# Generate random dataset
np.random.seed(0)  # For reproducibility
x = np.arange(100)  # X-values: 0 to 99
y = np.random.randn(100)  # Random Y-values from a normal distribution

# Create a line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with labels and title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
````
