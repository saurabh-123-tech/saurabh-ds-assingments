```python
#3 Write a code to check if a given string is a palindrome or not

def is_palindrome(s):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    s = s.replace(" ", "").lower()
    # Compare the string to its reverse
    return s == s[::-1]

# Example usage:
input_string = "A man a plan a canal Panama"
if is_palindrome(input_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
#4 Write a code to check if two given strings are anagrams of each other

def are_anagrams(s1, s2):
    # Remove spaces and convert to lowercase for case-insensitive comparison
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()
    
    # Check if the sorted characters of both strings are the same
    return sorted(s1) == sorted(s2)

# Example usage:
string1 = "Listen"
string2 = "Silent"
if are_anagrams(string1, string2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

```python
#5 Write a code to find all occurrences of a given substring within another string

def find_all_occurrences(s, substring):
    start = 0
    occurrences = []
    
    while True:
        start = s.find(substring, start)
        if start == -1:
            break
        occurrences.append(start)
        start += len(substring)  # Move past the current match
    
    return occurrences

# Example usage:
main_string = "The quick brown fox jumps over the lazy dog. The fox is quick."
substring = "fox"
indices = find_all_occurrences(main_string, substring)
print(f"Occurrences of '{substring}' found at indices: {indices}")
```

```python
#6 Write a code to perform basic string compression using the counts of repeated characters

def compress_string(s):
    if not s:
        return ""

    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last set of characters
    compressed.append(s[-1] + str(count))

    compressed_string = ''.join(compressed)
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage:
input_string = "aaabbccccd"
compressed_string = compress_string(input_string)
print(f"Compressed string: {compressed_string}")
```

```python
#7 Write a code to determine if a string has all unique characters

Approach 1: Using a Set
A straightforward method is to use a set to track characters and check for duplicates.

def has_unique_characters(s):
    # Use a set to track characters
    seen = set()
    
    for char in s:
        if char in seen:
            return False
        seen.add(char)
    
    return True

# Example usage:
input_string = "abcdefg"
if has_unique_characters(input_string):
    print("The string has all unique characters.")
else:
    print("The string does not have all unique characters.")

    
Approach 2: Sorting
Another approach is to sort the string and then check adjacent characters for duplicates.

def has_unique_characters(s):
    # Sort the string
    sorted_s = sorted(s)
    
    # Check for consecutive duplicates
    for i in range(len(sorted_s) - 1):
        if sorted_s[i] == sorted_s[i + 1]:
            return False
    
    return True

# Example usage:
input_string = "abcdefg"
if has_unique_characters(input_string):
    print("The string has all unique characters.")
else:
    print("The string does not have all unique characters.")

    
Approach 3: Using a Boolean Array
For a string with ASCII characters, you can use a boolean array to track characters. This approach assumes the string only contains ASCII characters.

def has_unique_characters(s):
    if len(s) > 256:  # ASCII character set size
        return False
    
    # Create a boolean array to track characters
    char_set = [False] * 256
    
    for char in s:
        index = ord(char)
        if char_set[index]:
            return False
        char_set[index] = True
    
    return True

# Example usage:
input_string = "abcdefg"
if has_unique_characters(input_string):
    print("The string has all unique characters.")
else:
    print("The string does not have all unique characters.")
```

```python
#12 Implement a code to find the second largest number in a given list of integers

def find_second_largest(numbers):
    if len(numbers) < 2:
        raise ValueError("List must contain at least two distinct numbers")

    # Initialize the largest and second largest with minimal values
    largest = second_largest = float('-inf')

    for num in numbers:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    if second_largest == float('-inf'):
        raise ValueError("No second largest number found, all numbers may be identical")

    return second_largest

# Example usage:
numbers_list = [10, 5, 8, 20, 20, 4, 7]
second_largest = find_second_largest(numbers_list)
print(f"The second largest number is: {second_largest}")
```

```python
#14 Write a code to reverse a list in-place without using any built-in reverse functions

def reverse_list_in_place(lst):
    left = 0
    right = len(lst) - 1
    
    while left < right:
        # Swap the elements at the left and right indices
        lst[left], lst[right] = lst[right], lst[left]
        # Move the pointers towards the center
        left += 1
        right -= 1

# Example usage:
my_list = [1, 2, 3, 4, 5]
reverse_list_in_place(my_list)
print(my_list)  # Output: [5, 4, 3, 2, 1]
```

```python
#15 Implement a code to find and remove duplicates from a list while preserving the original order of elements

def remove_duplicates(lst):
    seen = set()  # To keep track of seen elements
    result = []   # To store elements in the original order without duplicates

    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)

    return result

# Example usage:
my_list = [1, 2, 2, 3, 4, 4, 5, 1]
no_duplicates = remove_duplicates(my_list)
print(no_duplicates)  # Output: [1, 2, 3, 4, 5]
```

```python
#16 Create a code to check if a given list is sorted (either in ascending or descending order) or not

def is_sorted(lst):
    if len(lst) <= 1:
        return True  # A list with 0 or 1 element is considered sorted

    ascending = True
    descending = True

    # Check for ascending order
    for i in range(1, len(lst)):
        if lst[i] < lst[i - 1]:
            ascending = False
            break

    # Check for descending order
    for i in range(1, len(lst)):
        if lst[i] > lst[i - 1]:
            descending = False
            break

    return ascending or descending

# Example usage:
list1 = [1, 2, 3, 4, 5]
list2 = [5, 4, 3, 2, 1]
list3 = [1, 3, 2, 4, 5]
list4 = [1, 1, 1]

print(is_sorted(list1))  # Output: True (sorted in ascending order)
print(is_sorted(list2))  # Output: True (sorted in descending order)
print(is_sorted(list3))  # Output: False (not sorted)
print(is_sorted(list4))  # Output: True (sorted, all elements are the same)
```

```python
#17 Write a code to merge two sorted lists into a single sorted list

def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0  # Pointers for list1 and list2

    # Traverse both lists and merge them in sorted order
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append remaining elements of list1, if any
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Append remaining elements of list2, if any
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage:
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
merged = merge_sorted_lists(sorted_list1, sorted_list2)
print(merged)  # Output: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
#20 Write a code to shuffle a given list randomly without using any built-in shuffle functions

import random

def shuffle_list(lst):
    # Fisher-Yates shuffle algorithm
    for i in range(len(lst) - 1, 0, -1):
        # Pick a random index from 0 to i
        j = random.randint(0, i)
        # Swap the elements at indices i and j
        lst[i], lst[j] = lst[j], lst[i]

# Example usage:
my_list = [1, 2, 3, 4, 5]
shuffle_list(my_list)
print(my_list)  # Output: A shuffled version of the list, e.g., [3, 1, 4, 5, 2]
```

```python
#32 Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together

def merge_dictionaries(dict1, dict2):
    # Create a new dictionary to hold the merged results
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary

    # Iterate over items in the second dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            # If the key is already in the merged dictionary, add the values
            merged_dict[key] += value
        else:
            # If the key is not in the merged dictionary, add it
            merged_dict[key] = value

    return merged_dict

# Example usage:
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 40}

merged_dict = merge_dictionaries(dict1, dict2)
print(merged_dict)
```

```python
#33 Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None

def get_nested_value(d, keys):
    """
    Access a value in a nested dictionary based on a list of keys.
    
    :param d: The nested dictionary.
    :param keys: A list of keys to traverse the dictionary.
    :return: The value corresponding to the list of keys, or None if any key is not found.
    """
    # Start with the outermost dictionary
    current_value = d
    
    # Traverse through the keys
    for key in keys:
        # Check if the current_value is still a dictionary and if the key exists
        if isinstance(current_value, dict) and key in current_value:
            # Move to the next level in the dictionary
            current_value = current_value[key]
        else:
            # Return None if any key is not found or if current_value is not a dictionary
            return None
    
    return current_value

# Example usage:
nested_dict = {
    'a': {
        'b': {
            'c': 10
        }
    },
    'x': {
        'y': 20
    }
}

keys_to_access = ['a', 'b', 'c']
value = get_nested_value(nested_dict, keys_to_access)
print(value)  # Output: 10

# Example with a key that does not exist
keys_to_access_invalid = ['a', 'b', 'd']
value_invalid = get_nested_value(nested_dict, keys_to_access_invalid)
print(value_invalid)  # Output: None
```

```python
#34 Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order

def sort_dict_by_value(d, descending=False):
    """
    Sort a dictionary by its values.

    :param d: The dictionary to be sorted.
    :param descending: Whether to sort in descending order (default is False for ascending order).
    :return: A new dictionary sorted by its values.
    """
    # Sort the dictionary items based on values
    sorted_items = sorted(d.items(), key=lambda item: item[1], reverse=descending)
    
    # Convert the sorted items back to a dictionary
    sorted_dict = dict(sorted_items)
    
    return sorted_dict

# Example usage:
input_dict = {'apple': 10, 'banana': 5, 'cherry': 20, 'date': 15}

# Sort in ascending order
sorted_dict_ascending = sort_dict_by_value(input_dict)
print("Sorted dictionary (ascending):", sorted_dict_ascending)

# Sort in descending order
sorted_dict_descending = sort_dict_by_value(input_dict, descending=True)
print("Sorted dictionary (descending):", sorted_dict_descending)
```

```python
#35 Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

def invert_dict(d):
    """
    Invert a dictionary by swapping keys and values. If multiple keys have the same value,
    store the keys as a list in the inverted dictionary.
    
    :param d: The dictionary to be inverted.
    :return: A new dictionary with keys and values swapped.
    """
    inverted_dict = {}
    
    for key, value in d.items():
        if value in inverted_dict:
            # If the value already exists in the inverted dictionary, append the key to the list
            inverted_dict[value].append(key)
        else:
            # If the value does not exist, create a new entry with the key in a list
            inverted_dict[value] = [key]
    
    return inverted_dict

# Example usage:
original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3
}

inverted_dict = invert_dict(original_dict)
print(inverted_dict)
```
